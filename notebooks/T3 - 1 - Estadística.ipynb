## Coeficiente de Relación de Pearson

```python
Image(filename='resources/pearson.png')
```

```python
import pandas as pd
```

```python
data_ads = pd.read_csv('../datasets/ads/Advertising.csv')
```

```python
data_ads.head()
```

```python
len(data_ads)
```

```python
import numpy as np
```

```python
data_ads['CorrN'] = (data_ads['TV'] - np.mean(data_ads['TV'])) * (data_ads['Sales'] - np.mean(data_ads['Sales']))
```

```python
data_ads.head()
```

```python
data_ads['Corr1'] = (data_ads['TV'] - np.mean(data_ads['TV']))**2
```

```python
data_ads.head()
```

```python
data_ads['Corr2'] = (data_ads['Sales'] - np.mean(data_ads['Sales']))**2
```

```python
data_ads.head()
```

```python
corr_pearson = sum(data_ads['CorrN'])/np.sqrt(sum(data_ads['Corr1']) * sum(data_ads['Corr2']))
```

```python
corr_pearson
```

```python
def corr_coef(df, var1, var2):
    df['CorrN'] = (df[var1] - np.mean(df[var1])) * (df[var2] - np.mean(df[var2]))
    df['Corr1'] = (df[var1] - np.mean(df[var1]))**2
    df['Corr2'] = (df[var2] - np.mean(df[var2]))**2
    corr_pearson = sum(data_ads['CorrN'])/np.sqrt(sum(data_ads['Corr1']) * sum(data_ads['Corr2']))
    return corr_pearson
```

```python
corr_coef(data_ads, 'TV', 'Sales')
```

```python
cols = data_ads.columns.values
```

```python
cols
```

```python
for i in cols:
    for j in cols:
        print('{} + {} : {}'.format(i, j, corr_coef(data_ads, i, j)))
```
