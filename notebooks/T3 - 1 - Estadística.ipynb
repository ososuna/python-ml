## Medidas de Dispersión
* La varianza es una forma de medir cómo los valores se desplazan mucho o poco respecto de la media, por lo tanto, la varianza siempre es un número positivo.
* La desviación típica es la raíz cuadrada de la varianza.
* El coeficiente de variación mide la variabilidad relativa entre la media y la desviación típica.
