## Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
```

```python
gender = ['Male', 'Female']
income = ['Poor', 'Middle Class', 'Rich']
```

```python
n = 500
gender_data = []
income_data = []

for i in range(0, 500):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[:10]
```

```python
income_data[:10]
```

```python
# N(m, s) -> m + s * z
heigth = 160 + 30 + np.random.randn(n)
weight = 65 + 25 * np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income = 18000 + 3500 * np.random.randn(n)
```

```python
data = pd.DataFrame(
    {
        'Gender': gender_data,
        'Economic Status': income_data,
        'Height': heigth,
        'Weight': weight,
        'Age': age,
        'Income': income
    }
)
```

```python
data.head()
```

## Agrupación de datos

```python
grouped_gender = data.groupby('Gender')
```

```python
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
# Genera un dataframe con un grupo determinado
grouped_gender.get_group('Female').head()
```

```python
double_group = data.groupby(['Gender', 'Economic Status'])
```

```python
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

## Operaciones sobre datos agrupados
Significa aplicar una función a todos y cada uno de los grupos a la vez y obtener un resultado para ese grupo en particular.

```python
double_group.sum()
```

```python
double_group.mean()
```

```python
double_group.size()
```

```python
double_group.describe()
```

```python
grouped_income = double_group['Income']
grouped_income.describe()
```

```python
double_group.aggregate(
    {
        'Income': np.sum,
        'Age': np.min,
        'Height': np.std
    }
)
```

```python
double_group.aggregate(
    {
        'Age': np.mean,
        'Height': lambda h: np.mean(h)-np.std(h)
    }
)
```

```python
double_group.aggregate([np.sum, np.mean, np.std])
```

```python
double_group.aggregate([lambda x: np.mean(x) / np.std(x)])
```

## Filtrado de datos

```python
double_group['Age'].filter(lambda x: x.sum() > 2400)
```

## Transformación de variables

```python
zscore = lambda x: (x - x.mean()) / x.std()
```

```python
z_group = double_group.transform(zscore)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(z_group['Age'])
```

```python
fill_na_mean = lambda x: x.fillna(x.mean())
```

```python
double_group.transform(fill_na_mean).head()
```
