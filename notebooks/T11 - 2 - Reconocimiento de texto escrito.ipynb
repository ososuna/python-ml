## Construcción de la red neuronal

Tensorflow en lugar de ejecutar una operación única de forma muy pesada e independiente fuera de Python, Tensorflow lo que hace es describir un grafo interconectado de operaciones que se ejecutan todas fuera de Python.

Para describir las operaciones que va a llevar a cabo el sistema fuera de Python, se utilizan los placeholders.

y = softmax (W * x + b)

```python
dim_input = 784
n_categories = 10
```

```python
x = tf.placeholder(tf.float32, [None, dim_input]) # None = cualquier longitud
```

```python
W = tf.Variable(tf.zeros([dim_input, n_categories]))
b = tf.Variable(tf.zeros([n_categories]))
```

Una variable es un tensor que se puede modificar que vive en el grafo de Tensorflow para todas las opraciones interactuantes.

```python
softmax_args = tf.matmul(x, W) + b
y_hat = tf.nn.softmax(softmax_args)
```

## Entrenamiento de la red neuronal

```python
from IPython.display import display, Math, Latex
```

```python
display(Math(r"H_{y}(\hat{y}) = -\sum_{i} y_i' log(\hat{y_i})"))
```

```python
y_ = tf.placeholder(tf.float32, [None, 10])
```

```python
cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_ * tf.log(y_hat), reduction_indices=[1]))
```

```python
# tf.nn.softmax_cross_entropy_with_logits(softmax_argsm y_)
```

Tensorflow añadirá nuevas operaciones al grafo que implementarán propagación hacia atrás y gradiente descendiente de modo que con cada operación que ejecuten, ejecutará un paso de entrenamiento con gradiente descendiente con una modificación adicional de las variables de los pesos y el bias que reduzcan la pérdida (entropía cruzada).

```python
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(cross_entropy)
```

```python
session = tf.InteractiveSession()
```

```python
tf.global_variables_initializer().run()
```

```python
for _ in range(5000):
    batch_x, batch_y = mnist.train.next_batch(100)
    session.run(train_step, feed_dict={x:batch_x, y_:batch_y})
```

En cada etapa del bucle obtenemos una muestra de 100 datos aleatorios de nuestro dataset, a partir de aquí ejecutamos la función train_step que hemos definido suministrándole esas 100 imágenes aleatorias, que son sustituidas como placeholder de la variable x.
