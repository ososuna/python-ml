### Carga de datos a través de la función open (carga manual)
Se lee el dataset línea por línea ejecutando un bucle for. Borra cada uno de los trozos de la memoria una vez que hayan sido procesados. Es útil en datasets muy grandes.

```python
data3 = open(main_path + "/" + "customer-churn-model/Customer Churn Model.txt", "r")
```

#### Método next()
El método next() navega hacia la siguiente posición de memoria de la siguiente línea (después de la
cabecera en este caso).

#### Método strip()
El método strip() elimina los espacios en blanco que sobran tanto al inicio como al final de la línea.

#### Método split()
El método split divide una línea de texto por un delimitador.

```python
cols = data3.readline().strip().split(",")
n_cols = len(cols)
n_cols
```

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
main_dict
```

```python
for line in data3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1
print('El dataset tiene {} filas y {} columnas'.format(counter, n_cols))
```

```python
df3 = pd.DataFrame(main_dict)
df3.head()
```

#### Cuidado con el método open
Aunque en la clase anterior vimos un ejemplo del método open de Python donde en este caso sólo se utilizó el modo "r" (lectura) siempre es bueno recordar lo siguiente para evitar problemas con los ficheros:

Hay que tener cuidado al abrir un fichero en modo "w" (escritura) ya que esto hará que dicho fichero pierda su contenido (en caso de tenerlo y querer conservarlo). Si por algún motivo se quiere agregar contenido a este es recomendable abrirlo en modo "a" (append), lo cual posicionará el cursor al final y no borrará el contenido del mismo.

## Lectura y escritura de ficheros

```python
infile = main_path + "/" + "customer-churn-model/Customer Churn Model.txt"
outfile = main_path + "/" + "customer-churn-model/Tab Customer Churn Model.txt"
```

```python
with open(infile, 'r') as infile1:
    with open(outfile, 'w') as outfile1:
        for line in infile1:
            fields = line.split(',')
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
```

```python
df4 = pd.read_csv(outfile, sep="\t")
df4.head()
```

### Ejercicio: descargar y procesar datos desde una URL externa

```python
def downloadFromURL(url, filename, sep = ',', delim = '\n', encoding = 'utf-8', method = 'GET',
                   main_path = "C:/Users/Oswaldo/Desktop/Cursos/Curso Completo de Machine Learning Data Science en Python/python-ml/datasets"):
    # Primero importamos la librería y hacemos la conexión con la web de los datos
    import urllib3
    import csv
    http = urllib3.PoolManager()
    r = http.request(method, url)
    print('El estado de la respuesta es {}'.format(r.status))
    response = r.data

    # El objeto response contiene un string binario, lo convertimos a un string descodificándolo en UTF-8
    str_data = response.decode(encoding)

    # Dividimos el string en un array de filas, separándolo por intros
    lines = str_data.split(delim)

    # La primera línea contiene la cabecera, así que la extraemos
    col_names = lines[0].split(sep)

    # Generamos un diccionario vacío donde irá la información procesada desde la URL externa
    medals_dict = {}
    counter = 0
    for col in col_names:
        medals_dict[col] = []

    # Procesamos fila a fila la información para ir rellenando el diccionario con los datos
    for line in lines:
        # Nos saltamos la primera línea que es la que contiene la cabecera y ya tenemos procesada
        if counter > 0:
            # Dividimos cada string por las comas como elemento separador
            values = line.strip().split(sep)
            # Añadimos cada valor a su respectiva columna del diccionario
            for i in range(len(col_names)):
                medals_dict[col_names[i]].append(values[i])
        counter += 1

    print('El dataset tiene {} columnas y {} filas'.format(len(col_names), counter))

    # Convertimos el diccionario procesado a Data Frame y comprobamos que los datos son correctos
    df = pd.DataFrame(medals_dict)

    # Elegimos donde guardarlo (en la carpeta athletes es donde tiene más sentido por el contexto del análisis)
    # filename = 'athletes/downloaded_medals'
    fullpath = os.path.join(main_path, filename)

    # Lo guardamos en CSV, en JSON o en Excel según queramos
    df.to_csv(fullpath+'.csv')
    df.to_json(fullpath+'.json')
    df.to_excel(fullpath+'.xls')
    print('Los ficheros se han guardado correctamente en "{}"'.format(fullpath))
    
    return df

medals_df = downloadFromURL(medals_url, 'athletes/downloaded_medals')
medals_df.head()
```
